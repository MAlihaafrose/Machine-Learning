**One hot encoding using Loop** array creat kore categorical jotogulu ache sobgulo dibo array te

```python
df7.head()
```

```python
cols = ['Gender']
for col in cols:
  dummy = pd.get_dummies(df7[col],drop_first=True,prefix='Gender')
  df8 = pd.concat([df7,dummy],axis=1).drop(col,axis=1)
df8.head()
```
